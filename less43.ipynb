```python
from sklearn import datasets
from sklearn.feature_selection import VarianceThreshold, SelectKBest, f_classif, SelectFromModel
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA, TruncatedSVD
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
iris = datasets.load_iris()
iris
```

```python
#Статистичний метод фільтрації ознак VarianceTreshold
selector = VarianceThreshold(threshold=0.4).fit_transform(iris.data)
selector
```

```python
# Перевіряю вручну дисперсію по цьому стовпцю
column_1 = iris.data[:,1]
mean_value = column_1.mean()
disp = np.sum((column_1 - mean_value)**2) / (len(column_1))
disp
```

```python
disp2 = column_1.var()
disp2
```

```python
#Статистичний метод фільтрації ознак SelectKBest
X = iris.data
y = iris.target
selector2 = SelectKBest(f_classif, k = 3).fit_transform(X,y)
selector2
```

```python
#Відбір ознак на основі ваг у моделі
sfm_selector = SelectFromModel(estimator=LogisticRegression(),threshold = 0.2).fit(X,y)
selected_features = [iris.feature_names[i] for i in (sfm_selector.get_support(indices=True))]
selected_features
```

```python
#Метод головних компонент PCA
pca = PCA(n_components = 4)
selector3 = pca.fit_transform(X)
selector3
```

```python
explained_variance_ratio = pca.explained_variance_ratio_
explained_variance_ratio
```

```python
#тут дивимось долю поясненої дисперсії для всіх компонентів, щоб визначити n_components
sum_of_components = sum(explained_variance_ratio)
sum_of_components
```

```python
# Реалізація SVD
svd = TruncatedSVD(n_components=3)
selector4 = svd.fit_transform(X)
selector4
```

```python
# Лінійний дискримінантний аналіз
lda = LinearDiscriminantAnalysis()
new_data = lda.fit_transform(X,y)
new_data
```

```python
explained_variance_ratio1 = lda.explained_variance_ratio_
explained_variance_ratio1
```
